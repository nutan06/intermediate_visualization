```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

In this notebook I am using iris dataset for the drill of presenting same data in multiple ways with the help of seaborn.

```python
#import csv
iris = pd.read_csv("./Iris.csv")
df = pd.DataFrame(iris)
iris.head()
```

```python
iris.groupby(['Species','SepalLengthCm','SepalWidthCm'], as_index = False).mean()['Species'].value_counts()
```

### Univariable plots

## 1 *Swarm plot*

```python
sns.set(style="darkgrid")
sns.swarmplot(x='Species', y='PetalLengthCm', data=df);
```

## 2 *Strip plot*

```python
sns.stripplot(x='Species', y='PetalLengthCm', data=df);
```

## 3 *Violin Plot*

```python
sns.violinplot(x='Species', y='PetalLengthCm', data=df);
```

# Bivariate Plots

## Scatter Plot

```python
sns.lmplot('SepalLengthCm',
           'SepalWidthCm', 
           data=df, 
           fit_reg=True,
           scatter_kws={'alpha':0.4});
```

## Joint Plot

```python
sns.jointplot(x= 'PetalLengthCm', y= 'PetalWidthCm', data=iris);
```

## Box plot

```python
sns.boxplot(x='Species', y='PetalLengthCm', data=df);
```

## Pair Plot

```python
sns.pairplot(df, hue="Species", size=3);
```
